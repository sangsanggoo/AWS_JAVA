```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from sklearn import datasets

pd_hydrant = pd.read_csv('전국소방용수시설표준데이터.csv',encoding='cp949')
pd_parking = pd.read_csv('전국소방자동차전용구역표준데이터.csv',encoding='cp949')

print(pd_hydrant.columns,pd_hydrant.shape)
print(pd_parking.columns,pd_parking.shape)
select_columns = ['시도명','소재지도로명주소','위도','경도']
df1 = pd_hydrant[select_columns]
df2 = pd_parking[select_columns]

df_combined = pd.concat((df1,df2),keys=['hydrant','parking'])
#print(df_combined.shape)
df_combined

df_combined.reset_index(inplace=True)
df_combined.drop('level_1',axis=1 , inplace=True)
cname_list = list(df_combined.columns)
cname_list[0] = '종류'
df_combined.columns  = cname_list
df_combined
```
