```python
import requests
import numpy as np
x = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
```

```python
data=[]
for line in x.text.split('\n') :
    data.append(line.split(','))
for sample in data :
    if len(sample) <1 :
        continue
    for i,value in enumerate(sample[:4]):
            try:
                sample[i] = float(value)
            except :
                sample[i] = 0.0
print(sample)
```

```python
for sample in data :
    if len(sample) < 2 :
        continue
    if 'setosa' in sample[-1] :
        sample[-1] = 0.0
    elif 'versicolor' in sample[-1] :
        sample[-1] = 1.0
    elif 'virginica' in sample[-1] :
        sample[-1] = 2.0
print(data[0])
print(data[51])
print(data[101])
print(data)
```
