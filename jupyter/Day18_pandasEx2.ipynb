```python
import numpy as np
import pandas as pd

foodtruck_df = pd.read_csv('전국푸드트럭허가구역표준데이터.csv',encoding='cp949')
foodtruck_df.head()
```

```python
foodtruck_df['시도명'].value_counts() #그냥 다 숫자를 세줌 ㅁㅊ
```

```python
print(foodtruck_df['시도명'].describe())
print(foodtruck_df['푸드트럭운영대수'].describe())
print(foodtruck_df[['시도명','푸드트럭운영대수']].describe())
```

```python
sub_data = foodtruck_df[['시도명','푸드트럭운영대수']]
sub_data.groupby('시도명').sum()
sub_data.groupby('시도명').count()
sub_data.groupby('시도명').std()
sub_data.groupby('시도명').describe()
sub_data.groupby('시도명').aggregate([np.sum,np.mean,np.std])
sub_data.groupby('시도명').aggregate(lambda x : np.sum(x)*100)
```
