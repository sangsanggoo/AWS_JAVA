```python
# return
def unique_count(raw_list,sort = False) :
    raw_set = set(raw_list)
    raw_key = []
    raw_value = []
    #print(raw_list)
    for i in raw_set :
     #  print(i,raw_list.count(i))
        raw_key.append(i)
        raw_value.append(raw_list.count(i))
    d = dict(zip(raw_key,raw_value))
    if sort == True:
        d = dic(sorted(d.items(),key = lambda x : x[1]))
    return d
unique_count(data_list)
```
