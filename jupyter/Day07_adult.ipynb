```python
import math
def mean(num_list) :
    return (sum(num_list)/len(num_list))

def dev(num_list): # 편차 : 각 값에서 평균을 뺀값
    m = mean(num_list)
    return [x-m for x in num_list]

def var(num_list) : #분산 : 각 편차를 제곱을 해서 총 개수로 나누기
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d])/(n-1)

def stdev(num_list) : #표준 편차 
    return math.sqrt(var(num_list))

def covar(list_1, list_2): #공분산
    n = len(list_1)
    list_1_dev= dev(list_1)
    list_2_dev= dev(list_2)
    return sum(x*y for x,y in zip(list_1_dev,list_2_dev))/(n-1)
    
def corel(list_1,list_2) : #상관도 (-1 ~ 1)
    return covar(list_1,list_2)/(stdev(list_1)*stdev(list_2))
```

```python
import requests
x = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
pass1 =x.text.split('\n')
pass2 = []
for i in pass1 :
    pass2.append(i.split(','))
money_list = []
time_list = []
age_list = []
for i in pass2[:500] :
    age_list.append(int(i[0]))
    money_list.append(int(i[10]))
    time_list.append(int(i[12]))
print(age_list)
print(money_list)
print(time_list)
```

```python
print('age / capital gain 상관도 : ',corel(age_list,money_list))
print('hours-per-week / capital gain 상관도 : ',corel(time_list,money_list))
```
