```python
import numpy as np
import pandas as pd
import matplotlib as plot

pop_df = pd.read_csv('등록인구추이_20230223163325.csv', encoding = 'cp949')
pop_df=pop_df.drop(0,axis=0)
pop_df.columns = ['시점','세대수','인구_남','인구_여','한국인','외국인','인구밀도']
pop_df.head()
```

```python
pop_df.insert(3,'남/여 인구비율',pop_df['인구_남'].astype(np.float64)/pop_df['인구_여'].astype(np.float64))
pop_df.head()
```

```python
pop_df.insert(6,'한국인/외국인 인구비율',pop_df['한국인'].astype(np.float64)/pop_df['외국인'].astype(np.float64))
```
