```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
f_path = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font', family=font_name)

pd_data = pd.read_csv('데이터산업_시장규모_20230306153115.csv',encoding='cp949')
pd_data.head()
print(pd_data.columns)

x = pd_data['시점']
col_names = pd_data.columns
fig = plt.figure(layout='tight')
line_axe = fig.add_subplot(2,1,1)
line_axe.set(title='2015~2021 산업 규모',ylabel = '억원')
#line_axe.set_title('2015~2021 산업 규모')
#line_axe.set_ylabe('억원')
line_axe.plot(x,pd_data['데이터 처리 및 관리 솔루션 개발·공급업'],'ro--',x,pd_data['데이터 구축 및 컨설팅 서비스업'],'g^--',x,pd_data['데이터 판매 및 제공 서비스업'],'bv:')

pie2015_axe = fig.add_subplot(2,2,3)
pie2015_axe.pie(pd_data.iloc[0,1:])
pie2015_axe.set_title('2015년 구성비')


pie2021_axe = fig.add_subplot(2,2,4)
pie2021_axe.pie(pd_data.iloc[-1,1:],labels=list('abc'),autopct='%.2f')
pie2021_axe.set_title('2021년 구성비')
```
