```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot

pd_data = pd.read_csv('내국인출국교통수단별_20230302171121.csv',encoding='cp949',header=1)
pd_data.drop('계',axis =1,inplace= True)
pd_data.columns = ['시점','공항','항구']
pd_data

#공항, 항구 년중 월별 출국 인원 plot
_, axe = plot.subplots()
x = pd_data['시점']
axe.plot(x,pd_data['공항'],x,pd_data['항구'])



#2015부터 2022년까지 추이 보기
```

```python

#nomarlization / standardzation 적용, plot
gong_min = pd_data['공항'].min()
gong_max = pd_data['공항'].max()
pd_data['gong_norm'] = pd_data['공항'].apply(lambda x: (x-gong_min)/(gong_max-gong_min))


goo_min = pd_data['항구'].min()
goo_max = pd_data['항구'].max()
pd_data['goo_norm'] = pd_data['항구'].apply(lambda x: (x-goo_min)/(goo_max-goo_min))

_, axe = plot.subplots()
axe.plot(pd_data['gong_norm'])
axe.plot(pd_data['goo_norm'])
```

```python
# standardization: (x-x_mean)/x_std

gong_min = pd_data['공항'].mean()
gong_std = pd_data['공항'].std()
pd_data['gong_std'] = pd_data['공항'].apply(lambda x: (x-gong_min)/gong_std)

goo_mean = pd_data['항구'].mean()
goo_std = pd_data['항구'].std()
pd_data['goo_std'] = pd_data['항구'].apply(lambda x: (x-goo_mean)/goo_std)

_, axee = plot.subplots()
axee.plot(pd_data['gong_norm'])
axee.plot(pd_data['goo_norm'])
```
